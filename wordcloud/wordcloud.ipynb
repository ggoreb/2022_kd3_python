1. 워드클라우드 관련 모듈 가져오기

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

2. 데이터 가져오기

```python
import requests
res = requests.get('http://ggoreb.com/python/file/날개.txt')
res.encoding = None
text = res.text
```

3. 워드클라우드 외형으로 사용할 Mask 불러오기

```python
# mask = np.array(Image.open('bird.png'))

icon = Image.open('bird.png')
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon, icon)
mask = np.array(mask)
```

4. 워드클라우드 객체 생성

```python
wordcloud = WordCloud(
  font_path='c:/windows/fonts/gulim.ttc', 
  background_color='white',
  mask=mask).generate(text)
```

5. 워드클라우드 표시

```python
plt.figure(figsize=(6, 6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
